Far words = different meaning

###Important Points

BoW ignores word order

BoW does not understand meaning

It only counts how many times a word appears

##Disadvantages of Bag of Words (BoW)

* In Bag of Words, all words are treated equally. Even if it is repeated many times.

* The model cannot understand meaning (semantics) of words.

* Important and unimportant words may get same value.

* Word context and order are ignored.

### Main Problem

* BoW only counts frequency

* It cannot decide importance of words

* It cannot capture semantic difference


###Solution: TF-IDF

To solve this problem, we use TF-IDF (Term Frequency ‚Äì Inverse Document Frequency).

Why TF-IDF?

1. Gives higher weight to important words

1. Reduces weight of common words

3. Helps in better text classification

```python
data = ["This is a sentence",
        "This is my first sentence, this is second sentence"]
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer()  #convert lowercase first then count the frequency
x = cv.fit_transform(data)
```

```python
dir(cv)
```

```python
import pandas as pd
pd.DataFrame(x.toarray(), columns=cv.get_feature_names_out())
# Create a DataFrame where rows represent documents and columns represent word/ngram frequencies
```

```python
cv.vocabulary_
```

```python
cv = CountVectorizer(ngram_range=(1,2))
x = cv.fit_transform(data)
```

```python
pd.DataFrame(x.toarray(), columns=cv.get_feature_names_out())
```

```python
pd.DataFrame({"Word":cv.get_feature_names_out(), "Count":x.toarray().sum(axis=0)}).sort_values("Count", ascending=False)
```

```python
cv.vocabulary_
```

```python
len(cv.vocabulary_)
```

```python
cv = CountVectorizer(stop_words='english')
x = cv.fit_transform(data)
```

```python
pd.DataFrame({"Word":cv.get_feature_names_out(), "Count":x.toarray().sum(axis=0)}).sort_values("Count", ascending=False)
```
